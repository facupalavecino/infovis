```python
import pandas as pd
```

```python
df = pd.read_csv("2022-04__2023-04__1200-2200.csv")
```

```python
participants = pd.read_csv(
    f"/Users/facundopalavecino/Documents/Facu/ITBA/Cursadas/08 - Data Viz/VAST-challenge/VAST-Challenge-2022/Datasets/Attributes/Participants.csv"
)

participants.head()
```

```python
jobs = pd.read_csv(
    f"/Users/facundopalavecino/Documents/Facu/ITBA/Cursadas/08 - Data Viz/VAST-challenge/VAST-Challenge-2022/Datasets/Attributes/Jobs.csv"
)

jobs.head()
```

```python
employers = pd.read_csv(
    f"/Users/facundopalavecino/Documents/Facu/ITBA/Cursadas/08 - Data Viz/VAST-challenge/VAST-Challenge-2022/Datasets/Attributes/Employers.csv"
)

employers.head()
```

```python
a = df.join(jobs, on="jobId", rsuffix="_Jobs")

b = a.join(participants, on="participantId", rsuffix="_Participants")

c = b.join(employers, rsuffix="_Employers")

c.head()
```

```python
c.columns
```

### Appendear edad media de los participantes

```python
avg_age_by_employer = c.groupby("employerId")["age"].agg(mean="mean")
```

```python
avg_age_by_employer.index = avg_age_by_employer.index.map(int)

avg_age_by_employer.index.names = ["employerId"]

avg_age_by_employer = avg_age_by_employer.rename(columns={"mean": "age"})

avg_age_by_employer.head()
```

```python
employers.head()
```

```python
employers = employers.set_index("employerId").join(avg_age_by_employer)
```

### Appendear salario promedio

```python
avg_hourlyRate_by_employer = jobs.groupby("employerId")["hourlyRate"].agg(mean="mean")
```

```python

avg_hourlyRate_by_employer = avg_hourlyRate_by_employer.rename(columns={"mean": "hourlyRate"})
avg_hourlyRate_by_employer.head()
```

```python
employers = employers.join(avg_hourlyRate_by_employer)
```

### Guardar

```python
employers.to_csv("Employersv3.csv")
```
